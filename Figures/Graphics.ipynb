# Here is graphics for our task

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
import math
from matplotlib import rc
rc('text', usetex=True)
```

### Function

Here is following function:
$$
f(x) = \frac{\exp{\left( a^T x + b \right)}}{1 + \exp{\left( a^T x+ b \right)}} \left( c^T x + d \right)
$$

a, b, x are two-dimension vectors

```python
a = np.array([1, 7]) # specify a
b = 3 # specify b
c = np.array([3, 4])
d = 11

def f(x):
    return (math.e ** (b + a @ x))/(1 + math.e ** (b + a @ x )) * (c @ x + d)
```

### Function points selection

```python
# select border
x_left = (-10, -10)
x_right = (10, 10)
freq = 100

def func_grid():
    x_min = x_left
    x_max = x_right
    x_1 = np.linspace(x_min[0], x_max[0], freq)
    x_lin_1 = np.array([x_1 for i in range(freq)])
    x_2 = np.linspace(x_min[1], x_max[1], freq)
    x_lin_2 = np.array(list(zip(*[x_2 for i in range(freq)])))
    grid_res = np.array([[f(np.array([i,j])) for i in x_2] for j in x_1])
    return x_lin_1, x_lin_2, grid_res
```

### Plane points selection

```python
plt.figure(figsize=(15, 10))
ax = plt.subplot(projection='3d')
ax.plot_wireframe(*func_grid())
ax.set_title(r'Function plot for a = ({}, {}) , b = {}, c= ({}, {}) , d = {}'.format(a[0], a[1], b, c[0], c[1], d))
plt.grid(True)
plt.savefig('figure.png')
plt.show()
```
